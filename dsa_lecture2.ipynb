## Hash Map

### Two Sum

https://leetcode.com/problems/two-sum/description/


```python
# Brute force solution

def twoSum1(nums, target):
    l = len(nums)
    for i in range(l-1):
        for j in range(i+1, l):
            if nums[i] + nums[j] == target:
                return [i, j]


nums = [2,7,11,15]
target = 9

twoSum1(nums, target)
```

```python
# Solution using hash map

def twoSum2(nums, target):
    h = {}
    l = len(nums)
    for i in range(l):
        diff = target - nums[i]
        if diff in h:
            return [h[diff], i]
        h[nums[i]] = i
        print(h)

nums = [2,7,11,15]
target = 9

twoSum2(nums, target)
```

```python
nums = list(range(100000))
target = 150000

import time

tic = time.time()
print('twoSum1 output is ', twoSum1(nums, target))
toc = time.time()
print('Time taken by twoSum1 is', toc-tic)


tic = time.time()
print('twoSum2 output is ', twoSum2(nums, target))
toc = time.time()
print('Time taken by twoSum2 is', toc-tic)
```



*   Hash map solution is much faster than brute force solution, but uses extra storage!



### Longest Substring Without Repeating Characters

https://leetcode.com/problems/longest-substring-without-repeating-characters/description/


```python
# Brute force solution

def ls1(s):
    l = len(s)
    maxlen = 0
    for start in range(l):
        for end in range(start+1, l):
            sub = s[start:end]
            lsub = end-start
            if len(set(sub)) == lsub:
                maxlen = max(maxlen, lsub)
    return maxlen



s = "abcabcbb"
ls1(s)
```

```python
# Solution using hash map

def ls2(s):
    l = len(s)
    maxlen = 0
    start = 0
    lastseen = {}
    for end in range(l):
        cur = s[end]
        if cur in lastseen:
            start = max(start, lastseen[cur] + 1)
        lsub = end-start+1
        maxlen = max(maxlen, lsub)
        lastseen[cur] = end
        #print('start, end: ' + str(start) + ', ' + str(end) +  '. substring: ' +  s[start:(end+1)])
        #print(lastseen)
    return maxlen

#s = "abcabcbb"
s = "munisreenivaspydi"
ls2(s)
```

```python
s = "munisreenivaspydi"*100

tic = time.time()
print('ls1 output is ', ls1(s))
toc = time.time()
print('Time taken by ls1 is', toc-tic)


tic = time.time()
print('ls2 output is ', ls2(s))
toc = time.time()
print('Time taken by ls2 is', toc-tic)
```

## Stack





*   We don't have an in-built `stack` data structure in Python, so we implement it ourselves.



```python
# Implementing a stack in python using a list

class Stack:
    def __init__(self):
        self.items = []

    def __str__(self):
        return '||' + str(self.items) + '>>'

    def is_empty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[-1]

    def size(self):
        return len(self.items)

browse = Stack()

browse.push('Napoleon Bonaparte')
browse.push('Louisiana Purchase')
browse.push('Mississipi River')
browse.pop()
browse.push('Luisiana')
print(browse)
```

### Valid Parentheses

https://leetcode.com/problems/valid-parentheses/

```python
def isvalid(s):
    bracketmap = {')':'(', ']':'[', '}':'{'}
    brackets = Stack()

    for c in s:
        if c in bracketmap:     # if c is a closing bracket
            if brackets.is_empty():
                return False
            elif brackets.peek() == bracketmap[c]:
                brackets.pop()
            else:
                brackets.push(c)
        else:                   # if c is an opening bracket
            brackets.push(c)
        print(c, brackets)

    if brackets.is_empty():
        return True
    else:
        return False

s = '(])'
isvalid(s)
```

### Daily Temperatures

https://leetcode.com/problems/daily-temperatures/description/

```python
# Brute force solution

def dt1(temps):
    l = len(temps)
    waits = [0]*l
    for id, temp in enumerate(temps[:-1]):
        idnext = id+1
        while (idnext <= (l-1)) and (temps[idnext] <= temp):
            idnext = idnext + 1
        if idnext <= l-1:
            waits[id] = idnext - id
    return waits

temps = [73,74,75,71,69,72,76,73]
dt1(temps)
```

```python
# Solution using a stack

def dt2(temps):
    tempstack = Stack()
    waits = [0]*len(temps)
    for id, temp in enumerate(temps):
        while (not tempstack.is_empty()) and temps[tempstack.peek()] < temp:
            tempind = tempstack.pop()
            waits[tempind] = id - tempind
        tempstack.push(id)
        print(tempstack)
    return waits

temps = [73,74,75,71,69,72,76,73]
dt2(temps)
```

```python
temps = [73,74,75,71,69,72,76,73]*1000

tic = time.time()
print('dt1 output is ', dt1(temps))
toc = time.time()
print('Time taken by dt1 is', toc-tic)


tic = time.time()
print('dt2 output is ', dt2(temps))
toc = time.time()
print('Time taken by dt2 is', toc-tic)
```

## Queue



```python
# Implementing a queue in python using a list

class Queue:
    def __init__(self):
        self.items = []

    def __str__(self):
        return '>>' + str(self.items) + '>>'

    def is_empty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0,item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

tasks = Queue()
tasks.enqueue('brush teeth')
tasks.enqueue('take shower')
tasks.dequeue()
tasks.enqueue('have breakfast')
print(tasks)
```

### Time Needed to Buy Tickets

https://leetcode.com/problems/time-needed-to-buy-tickets/description/

```python
def tn1(tickets, k):
    line = Queue()
    for id, num in enumerate(tickets):
        line.enqueue([id, num])

    time = 0
    while not line.is_empty():
        #print(line)
        time += 1
        id, num = line.dequeue()
        if (id == k) and (num == 1):
            return time
        if num > 1:
            line.enqueue([id, num-1])

tickets = [2,3,2]
k = 2
tn1(tickets, k)
```

### Efficient Queue using Deque

The Deque data structure in python implements a double-ended queue, which can be used as a stack or a queue.

```python
from collections import deque

groceries = deque(['bread', 'milk', 'eggs'])

print(type(groceries))
```

```python
groceries.append('juice')       # add a new item to the end of the deque
print(groceries)

groceries.appendleft('apples')  # add a new item to the start of the deque
print(groceries)

groceries.pop()                 # remove an item from the end of the deque
print(groceries)

groceries.popleft()             # remove an item from the start of the deque
print(groceries)
```

```python
# Implementing an efficient queue in python using a deque

from collections import deque

class EQueue:
    def __init__(self):
        self.items = deque()

    def __str__(self):
        return '>>' + str(self.items) + '>>'

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.appendleft(item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

tasks = Queue()
tasks.enqueue('brush teeth')
tasks.enqueue('take shower')
tasks.dequeue()
tasks.enqueue('have breakfast')
print(tasks)
```

```python
def tn2(tickets, k):
    line = EQueue()
    for id, num in enumerate(tickets):
        line.enqueue([id, num])

    time = 0
    while not line.is_empty():
        #print(line)
        time += 1
        id, num = line.dequeue()
        if (id == k) and (num == 1):
            return time
        if num > 1:
            line.enqueue([id, num-1])

tickets = [2,3,2]
k = 2
tn2(tickets, k)
```

```python
tickets = [2, 3, 5, 2, 1, 6]*10000
k = 50000

tic = time.time()
print('tn1 output is ', tn1(tickets, k))
toc = time.time()
print('Time taken by tn1 is', toc-tic)


tic = time.time()
print('tn2 output is ', tn2(tickets, k))
toc = time.time()
print('Time taken by tn2 is', toc-tic)
```
